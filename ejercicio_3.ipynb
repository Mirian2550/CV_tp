```python
import gdown
import os
import shutil


def download_from_drive(url, output_dir):
    """
    Downloads the folder from Google Drive to the specified output directory.
    If the directory already exists, it will be removed and replaced by the new download.

    Parameters:
    - url (str): The URL of the Google Drive folder.
    - output_dir (str): The directory where the folder will be downloaded.
    """
    if os.path.exists(output_dir):
        shutil.rmtree(output_dir)
    os.makedirs(output_dir)
    gdown.download_folder(url, output=output_dir, quiet=False)
    print(f"Download completed and replaced the folder: {output_dir}.")

url_other = 'https://drive.google.com/drive/folders/15vMYnhA1zBOn2Ua69O01ZdbnpkMYLHFn?usp=sharing'
output_dir_other = 'fotos_frutas'
download_from_drive(url_other, output_dir_other)

url_videos = 'https://drive.google.com/drive/folders/1WMS-g-gzp-ku1HGs9AW3EonUSBVWb1fJ'
output_dir_videos = 'videos_frutas'
download_from_drive(url_videos, output_dir_videos)

url_other = 'https://drive.google.com/drive/u/1/folders/1_8HTzIySFfDBcB_quHgIL7w3AVd6fD0F'
output_dir_other = '_modelos'
download_from_drive(url_other, output_dir_other)
```

## Instalamos Dependencias

```python
import cv2
import matplotlib.pyplot as plt
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo
import numpy as np
from pydantic import BaseModel, Field
import supervision as sv
```

El enfoque aplicado para resolver el problema de detección y conteo de frutas se basa en el uso de Detectron2, un potente modelo preentrenado que hemos ajustado (fine-tuning) para detectar correctamente nuestras clases específicas de frutas. Implementamos una clase modular, FruitDetector, que gestiona la detección y el conteo de frutas mediante un análisis por umbrales para asegurar que solo se cuenten las detecciones con alta confianza. Para los videos y flujos en tiempo real, integramos ByteTrack, un algoritmo de seguimiento que evita el conteo repetido de las mismas frutas en distintos frames. Además, el conteo acumulativo se muestra visualmente junto con las cajas de detección, ofreciendo una solución eficiente y precisa.

```python
class Config(BaseModel):
    """
    Configuración del modelo para la detección de frutas.

    Attributes:
        model (str): Ruta a los pesos del modelo preentrenado.
        class_names (list[str]): Lista de nombres de las clases de frutas.
        thresholds_per_class (dict): Diccionario con los umbrales de confianza por clase.
        device (str): Dispositivo en el que se ejecuta el modelo ('cpu' o 'cuda').
    """
    model: str = Field(..., description="Ruta a los pesos del modelo")
    class_names: list[str] = Field(..., description="Nombres de las clases de frutas")
    thresholds_per_class: dict = Field(..., description="Umbrales de confianza por clase")
    device: str = Field(default='cpu', description="Dispositivo a utilizar ('cpu' o 'cuda')")
```

```python
class FruitDetector:
    """
    Clase para detectar y contar frutas en imágenes, videos y cámara en tiempo real.

    Attributes:
        cfg (CfgNode): Configuración del modelo Detectron2.
        predictor (DefaultPredictor): Predictor de Detectron2 para realizar inferencias.
        tracker (ByteTrack): Algoritmo ByteTrack para el seguimiento de objetos en video.
        total_fruit_counts (dict): Conteo acumulado de frutas detectadas.
        counted_ids (set): Conjunto de IDs de objetos ya contados para evitar duplicados.
    """

    def __init__(self, config: Config):
        """
        Inicializa la clase con el modelo preentrenado, nombres de las clases y umbrales por clase.

        Args:
            config (Config): Configuración del modelo y parámetros.
        """
        self.class_names = config.class_names
        self.thresholds_per_class = config.thresholds_per_class

        self.cfg = get_cfg()
        self.cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
        self.cfg.MODEL.WEIGHTS = config.model  # Cambio de model_weights a model
        self.cfg.MODEL.ROI_HEADS.NUM_CLASSES = len(config.class_names)
        self.cfg.MODEL.DEVICE = config.device

        self.predictor = DefaultPredictor(self.cfg)
        self.tracker = sv.ByteTrack()
        self.total_fruit_counts = {class_name: 0 for class_name in self.class_names}
        self.counted_ids = set()

    def process_frame(self, frame: np.ndarray) -> np.ndarray:
        """
        Procesa un frame de imagen para detectar y contar frutas.

        Args:
            frame (np.ndarray): Imagen o frame a procesar.

        Returns:
            np.ndarray: Frame procesado con cajas de detección y conteo de frutas.
        """
        outputs = self.predictor(frame)
        instances = outputs["instances"].to("cpu")
        pred_boxes = instances.pred_boxes if instances.has("pred_boxes") else None
        pred_classes = instances.pred_classes if instances.has("pred_classes") else None
        scores = instances.scores if instances.has("scores") else None

        current_frame_counts = {class_name: 0 for class_name in self.class_names}

        if pred_boxes is not None:
            for i, box in enumerate(pred_boxes):
                class_id = pred_classes[i].item()
                class_name = self.class_names[class_id]
                class_threshold = self.thresholds_per_class[class_name]

                if scores[i] > class_threshold:
                    current_frame_counts[class_name] += 1
                    box = box.numpy()
                    start_point = (int(box[0]), int(box[1]))
                    end_point = (int(box[2]), int(box[3]))
                    frame = cv2.rectangle(frame, start_point, end_point, color=(0, 255, 0), thickness=4)

                    label = f"{class_name}: {scores[i].item():.2f}"
                    cv2.putText(frame, label, (int(box[0]), int(box[1]) - 20),
                                cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 255, 0), 4)

        for class_name in self.class_names:
            self.total_fruit_counts[class_name] += current_frame_counts[class_name]

        info_text = ' | '.join([f"{cls}: {count}" for cls, count in self.total_fruit_counts.items()])
        cv2.putText(frame, info_text, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2.0, (255, 0, 0), 6)

        return frame

    def process_image(self, image_path: str) -> None:
        """
        Procesa una imagen para detectar frutas.

        Args:
            image_path (str): Ruta a la imagen que se va a procesar.
        """
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"No se pudo cargar la imagen en {image_path}")

        processed_frame = self.process_frame(image)

        plt.figure(figsize=(10, 10))
        plt.imshow(cv2.cvtColor(processed_frame, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

    def process_frame_video(self, frame: np.ndarray) -> np.ndarray:
        """
        Procesa un frame de video para detectar y contar frutas.

        Args:
            frame (np.ndarray): Frame del video.

        Returns:
            np.ndarray: Frame procesado con cajas de seguimiento y conteo de frutas.
        """
        outputs = self.predictor(frame)
        instances = outputs["instances"].to("cpu")
        pred_boxes = instances.pred_boxes if instances.has("pred_boxes") else None
        pred_classes = instances.pred_classes if instances.has("pred_classes") else None
        scores = instances.scores if instances.has("scores") else None

        boxes, confidences, class_ids = [], [], []

        if pred_boxes is not None:
            for i, box in enumerate(pred_boxes):
                class_id = pred_classes[i].item()
                class_name = self.class_names[class_id]
                score = scores[i].item()
                class_threshold = self.thresholds_per_class[class_name]

                if score > class_threshold:
                    bbox = [float(box[0].item()), float(box[1].item()), float(box[2].item()), float(box[3].item())]
                    boxes.append(bbox)
                    confidences.append(score)
                    class_ids.append(class_id)

        if len(boxes) > 0:
            detections = sv.Detections(
                xyxy=np.array(boxes),
                confidence=np.array(confidences),
                class_id=np.array(class_ids)
            )
            tracks = self.tracker.update_with_detections(detections)

            for track in tracks:
                bbox, _, score, class_id, track_id, _ = track
                class_name = self.class_names[class_id]
                x1, y1, x2, y2 = map(int, bbox)
                frame = cv2.rectangle(frame, (x1, y1), (x2, y2), color=(0, 255, 0), thickness=2)

                label = f"{class_name} ID: {track_id} Score: {score:.2f}"
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 255, 0), 5)

                if track_id not in self.counted_ids:
                    self.total_fruit_counts[class_name] += 1
                    self.counted_ids.add(track_id)

            info_text = ' | '.join([f"{cls}: {count}" for cls, count in self.total_fruit_counts.items()])
            cv2.putText(frame, info_text, (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 3, (255, 0, 0), 6)

        return frame

    def process_video(self, video_path: str) -> None:
        """
        Procesa un video para detectar y contar frutas.
    
        Args:
            video_path (str): Ruta al archivo de video.
        """
        cap = cv2.VideoCapture(video_path)
        if not cap.isOpened():
            raise ValueError(f"No se pudo abrir el video {video_path}")
    
        frame_count = 0
        processed_frame = None
    
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
    
            # Procesar cada 10 frames
            if frame_count % 10 == 0:
                processed_frame = self.process_frame_video(frame)
    
            # Mostrar el frame procesado o el frame normal
            cv2.imshow('Detección de Frutas', processed_frame if processed_frame is not None else frame)
            
            # Salir si se presiona la tecla 'q'
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    
            frame_count += 1
    
        # Liberar recursos
        cap.release()
        cv2.destroyAllWindows()

    def process_camera(self) -> None:
        """
        Procesa el feed de una cámara en tiempo real y realiza detecciones de frutas.

        """
        cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            raise ValueError("No se pudo acceder a la cámara")

        frame_count = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if frame_count % 40 == 0:
                processed_frame = self.process_frame(frame)
                cv2.imshow('Detección de Frutas (Cámara)', processed_frame)
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break

            frame_count += 1

        cap.release()
        cv2.imshow('Detección de Frutas (Último Frame)', processed_frame)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
